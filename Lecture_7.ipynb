# Downloading Historical Stock Data

## Single Ticker

To download historical data for a single ticker symbol, use the download function:

```python
data = yf.download('AAPL', start='2022-01-01', end='2022-12-31')
```

## Multiple Tickers

You can also download data for multiple tickers simultaneously:

```python
tickers = ['AAPL', 'MSFT', 'GOOGL']
data = yf.download(tickers, start='2022-01-01', end='2022-12-31')
```

# Data Manipulation and Analysis

The downloaded data is returned as a Pandas DataFrame, which makes it easy to manipulate and analyze.

```python
data.head()
```

## Accessing Specific Columns

You can access specific columns like ‘Close’ prices:

```python
close_prices = data['Close']
```

## Calculating Daily Returns

```python
daily_returns = close_prices.pct_change()
```

# Downloading Other Types of Data

## Dividends and Splits

You can download dividends and stock splits data:

```python
apple = yf.Ticker('AAPL')
dividends = apple.dividends
splits = apple.splits
```

## Financial Statements

Retrieve financial statements like balance sheets, income statements, and cash flow statements:

```python
balance_sheet = apple.balance_sheet
income_statement = apple.financials
cash_flow = apple.cashflow
```
